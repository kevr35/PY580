```python
def merge_stock_coin(stock, coin, start_date):

    idx = yf.Ticker(stock)
    company = idx.history(stock,start = start_date)
    company = company.reset_index()
    company['Date'] = company['Date'].astype(str)

    crypto = pd.read_csv(r'./crypto_history/BTC-USD.csv')

    # using merge function by setting how='inner'
    merged_df = pd.merge(crypto,company, 
                       on='Date', 
                       how='inner')

    #merged_df_crypto = merged_df[["Date","Close_x"]]
    merged_df_stock = merged_df[["Date","Close_y"]]

    #merged_df_crypto = merged_df_crypto.rename(columns={"Close_x": coin})
    merged_df_stock = merged_df_stock.rename(columns={"Close_y": stock})
    merged_df_stock = merged_df_stock.set_index('Date')
    merged_df_stock = merged_df_stock.transpose()
    
    return merged_df_stock
  
# displaying result
msft = merge_stock_coin("MSFT", "BTC", "2017-12-01")    
msft
```
